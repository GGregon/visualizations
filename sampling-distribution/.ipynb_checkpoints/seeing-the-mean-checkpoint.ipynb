```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
# %config InlineBackend.figure_format = 'retina'
```

# Primeira análise

* Sintetizar dados de uma população com $N$ elementos segundo uma distruição normal
* Obter duas amostras de tamanho $n$, a partir dessa população - as amostras devem ter mesma distribuição marginal
* Exibir um gráfico 3D com os dados (dispersão) no plano $z=0$ e com a pdf conjunta. Nos planos, projetar a pdf marginal e num segundo gráfico, essas pdf's marginais sobrepostas

## Modelando a população $\mathcal{N}(\mu,\sigma)$

```python
# Parâmetros da população
pop_loc = 175
pop_scale = 9

# Obtemos os percentis máximos e mínimos
P_MIN=0.001
P_MAX=1-P_MIN

# Calculamos os quantis inversos para .1 e .999.
inv_q_min = norm.ppf(q=P_MIN,loc=pop_loc,scale=pop_scale)
inv_q_max = norm.ppf(q=P_MAX,loc=pop_loc,scale=pop_scale)
```

```python
POP_SIZE = 10000
x = norm.rvs(loc=pop_loc,scale=pop_scale,size=POP_SIZE)
```

```python
# Criamos um intervalo uniforme de valores para plotar a distribuição
x_vals = np.linspace(inv_q_min,inv_q_max,POP_SIZE)

# Calculamos as probabilidades de cada valor de x_vals
pdf_x_vals = norm.pdf(x_vals,loc=pop_loc,scale=pop_scale)
```

## Visualizando a população

```python
jitter = np.random.normal(0,0.01,POP_SIZE)
f,(ax1,ax2) = plt.subplots(2,1,height_ratios=[3,1],sharex=True)
ax1.plot(x_vals, pdf_x_vals,color='k')
ax2.scatter(x,jitter,alpha=0.2,s=1,color='k')
ax2.axis('off')
plt.show()
```

## Amostrando da população

Vamos considerar duas amostras de tamanho $n$

```python
n=60
s1 = np.random.choice(x,size=n)
s2 = np.random.choice(x,size=n)
```

Agora irei exibir os dados amostrados e suas distribuições no mesmo gráfico

```python
# Calculamos as probabilidades de cada valor de x_vals
pdf_s1 = norm.pdf(np.sort(s1),loc=np.mean(s1),scale=np.std(s1))
pdf_s2 = norm.pdf(np.sort(s2),loc=np.mean(s2),scale=np.std(s2))
```

```python
def jitter(n,loc=0,scale=0.01):
    return np.random.normal(loc,scale,size=n)
```

```python
f,(ax1,ax2) = plt.subplots(2,1,height_ratios=[3,1],sharex=True)
f.subplots_adjust(hspace=0.05)

ax1.hist(np.sort(s1), density=True, histtype='step',label='$X_1$',color='tab:red')
ax1.hist(np.sort(s2), density=True, histtype='step',label='$X_2$',color='tab:blue')
ax1.plot(x_vals, pdf_x_vals,label='População',c='k',alpha=0.3,ls='--')
ax1.legend()

ax2.scatter(s1,jitter(n),alpha=0.5,s=10,c='tab:red')
ax2.scatter(s2,jitter(n,0.02),alpha=0.5,s=10,c='tab:blue')
ax2.axis('off')
plt.show()
```

As amostras acima são independentes e aproximadamente igualmente distribuidas, pois foram obtidas a partir da mesma população aleatoriamente. Assim, vamos esboçar o gráfico 3D da sua distribuição conjunta

```python
S1, S2 = np.meshgrid(np.sort(s1),np.sort(s2))
```

```python
joint_pdf = np.outer(pdf_s1,pdf_s2)
```

```python
fig = plt.figure(figsize=(15, 5))

# Plot the surface
colormap_name='coolwarm'
ax3d = fig.add_subplot(121, projection='3d')
ax3d.contour(S1, S2, joint_pdf, zdir='x', levels=5, offset=S1.max(), cmap='Blues')
ax3d.contour(S1, S2, joint_pdf, zdir='y', levels=5, offset=S2.max(), cmap='Reds')
ax3d.scatter(S1[0], S2[:,0], 0,color='k',alpha=0.1)

ax2d = fig.add_subplot(122)
ax2d.plot(np.sort(s1),pdf_s1,color='tab:red')
ax2d.plot(np.sort(s2),pdf_s2,color='tab:blue')


for ax in ax.axis
ax.xaxis.pane.fill = False   # Remove color on the XZ plane
ax.yaxis.pane.fill = False   # Remove color on the YZ plane
ax.zaxis.pane.fill = False   # Remove color on the XY plane
ax.set_zticklabels([])
# ax.zaxis.set_ticks_position('lower')   # 'left', 'right', 'top', 'bottom', 'none'
# ax.zaxis.set_label_position('lower')   # 'left', 'right', 'top', 'bottom'

ax.xaxis._axinfo['grid'].update(color = 'whitesmoke')
ax.yaxis._axinfo['grid'].update(color = 'whitesmoke')
ax.zaxis._axinfo['grid'].update(color = 'whitesmoke')

ax.set(xlim=(S1.min(),S1.max()),ylim=(S2.min(),S2.max()))
ax.set_xlabel('X1')
ax.set_ylabel('X2')

ax.grid(alpha=0.3)
ax.view_init(elev=30, azim=-120)
plt.show()
```
