## Descrição

Neste notebook, irei resolver um problema envolvendo tempo de espera definido por dois agentes, cada qual com sua distribuição de tempo de chegada ao local definido. Após resolver o caso simples, irei estressar o conceito e analisar outros casos para distribuições diferentes. Para isso, vou trabalhar sob o seguinte enunciado:

> _Duas pessoas, A e B, combinam de se encontrar entre 1 e 2 da tarde num determinado local. Considerando que eles chegam ao local de modo independente e aleatoriamente ao longo do intervalo de uma hora, descubra a distribuição do tempo que A aguarda B. (Considere que se B chega antes de A o tempo de espera é zero)_


## Configurações e ambiente

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import norm
from scipy.stats import gaussian_kde

# %config InlineBackend.figure_format = 'retina'
```

Vou criar dados sintéticos que representam os tempos em que cada pessoa chega ao local combinado. Inicialmente, irei considerar que esses tempos seguem distribuições normais

```python
# Janela de tempo (em minutos)
x = np.linspace(0, 60, 500)

# Características da distribuição de cada pessoa
media_A = 15
media_B = 15
desvio_A = 5
desvio_B = 2
```

```python
# Função que retorna os valores de x e y para plotar a pdf
def normal_data_from_pdf(loc,scale,N):
    x_min = norm.ppf(0.001,loc=loc,scale=scale)
    x_max = norm.ppf(1-0.001,loc=loc,scale=scale)
    x_vals = np.linspace(x_min,x_max,N)
    pdf_vals = norm.pdf(x_vals,loc=loc,scale=scale)
    return x_vals, pdf_vals
```
